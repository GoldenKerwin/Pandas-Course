```python
fig = plt.figure()   #通过 plt.figure() 创建一个空白的图形窗口
ax1 = fig.add_subplot(2,2,1)  #使用 fig.add_subplot(2,2,1)，将图形分割为2行2列的子图区域，并选择将第一个子图定义为当前操作的子图
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(),'k--')   #'k--'是用于绘制黑色分段线的style选项
```

```python
_ = ax1.hist(np.random.randn(100), bins=20, color='k',alpha=0.3)
#hist表示绘制了一个直方图。np.random.randn(100) 生成了一个包含100个随机数的数组，bins=20 指定了直方图的条柱数量为20个，
#color='k' 设置直方图的颜色为黑色，alpha=0.3 设置直方图的透明度为0.3。
#函数返回一个元组，其中包含直方图的数值、边界和条柱对象。使用 _ 表示我们不关心返回值，仅仅将直方图绘制在子图中
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
#scatter绘制了一个散点图。np.arange(30) 生成了一个包含30个等差数列的数组，
#np.arange(30) + 3 * np.random.randn(30) 则生成了一个与前者相加后加上一些噪声的数组。
#通过 scatter 函数，将这两个数组作为输入，在子图中绘制了相应的散点图。
fig
```

```python
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax, style='k-')
crisis_data = [
    (datetime(2007,10,11), 'Peak of bull market'),   #牛市顶峰
    (datetime(2008,3,12), 'Bear Stearns Fails'),     #贝尔斯登倒闭
    (datetime(2008,9,15), 'Lehman Bankruptcy')       #雷曼破产
]      #定义了三个重要日期的元组列表，每个元组包含日期和相应的标注文本。

for date, label in crisis_data:    #通过循环遍历危机日期列表
    ax.annotate(label,xy=(date, spx.asof(date)+75),   #xy: 注释箭头的目标位置，即要标注的点的位置。
                                                      #调用spx.asof(date)时，它会返回最接近给定日期的标普500指数的值
               xytext=(date, spx.asof(date)+225),     #xytext: 注释文本的位置，即文本所在的坐标位置
               arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4),
               horizontalalignment='left', verticalalignment='top')
#arrowprops: 用于设置箭头的样式，箭头的颜色为黑色，箭头头部的宽度为4，箭身的宽度为2，箭头的长度为4
# horizontalalignment: 文本水平对齐方式，这里设置为左对齐
# verticalalignment: 文本垂直对齐方式，这里设置为顶部对齐
#使用annotate函数在图上添加注释。xy参数指定注释的位置，xytext参数指定注释文本的位置，arrowprops参数指定箭头的样式

ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([600,1800])    #设置图表的X轴和Y轴的显示范围
ax.set_title('Important dates in the 2008-2009 financial crisis')   #设置图表的标题
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3)
#Rectangle()函数创建一个宽度为0.4，高度为0.15的黑色矩形，左下角坐标为(0.2, 0.75)，透明度（alpha）设置为0.3
circ = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)
#Circle()函数创建一个半径为0.15的蓝色圆，中心坐标为(0.7, 0.2)，透明度设置为0.3
pgon = plt.Polygon([[0.15,0.15], [0.35,0.4], [0.2,0.6]],
                  color='g',alpha=0.5)
#Polygon()函数创建一个三角形，顶点坐标分别为(0.15, 0.15)、(0.35, 0.4)和(0.2, 0.6)，颜色为绿色，透明度设置为0.5
ax.add_patch(rect)    #通过ax.add_patch()方法将这三个形状添加到子图中
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
tips = pd.read_csv('examples/tips.csv')
party_counts = pd.crosstab(tips['day'],tips['size'])
#pd.crosstab()函数将这两个变量作为参数，生成一个包含交叉频数的新的DataFrame对象，将其赋值给party_counts变量
party_counts
```

```python
party_counts = party_counts.loc[:, 2:5]
#loc属性对party_counts进行了列筛选，选择了包含顾客人数在2~5人之间的四列数据, loc直接对的是标签，不是位置数
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
#div()函数将party_counts中的每个值除以该行总和， 设置axis参数为0以沿着行方向执行操作
#该行代码最终将结果赋值给了party_pcts变量，计算了每个顾客人数组合的百分比
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
import seaborn as sns
tips['tip_pct'] = tips['tip']/(tips['total_bill']-tips['tip'])
tips
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
#x='tip_pct'：条形图上的横轴数据是名为"tip_pct"的变量。这表示条形的长度或高度
#y='day'：条形图上的纵轴数据是名为"day"的变量。这表示条形的位置或分组
#data=tips：绘图所使用的数据来自于名为"tips"的DataFrame对象。
#orient='h'：设置条形图为水平方向。即条形水平放置，纵轴表示分组，横轴表示数值
#柱子的值是tip_pct的平均值，黑线代表95%的置信区间，
#置信区间是用来估计总体参数范围的统计概念，提供了总体参数估计的不确定性信息，并通常与置信水平相关，以表示对这个估计结果的可信程度
```

```python
a = sns.barplot(x='tip_pct',y='day',hue='time',data=tips,orient='h')
a = sns.set(style='whitegrid')
#hue选项允许我们通过一个额外的分类值将数据分离
```

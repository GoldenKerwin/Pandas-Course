```python
chunker = pd.read_csv('examples/ex6.csv',chunksize=1000)    #参数表示以每块1000行的大小分块读取文件，以节省内存。
tot = pd.Series([])
for piece in chunker:                  #使用 for 循环遍历 chunker，每次迭代都会得到一个数据块 piece
    tot = tot.add(piece['key'].value_counts(),fill_value=0) #在每个数据块 piece 中，使用 ['key'].value_counts() 统计 'key' 列中各个值出现的频数
                                                            #并将结果与 tot 进行累加操作。fill_value=0 参数表示缺失值用 0 填充
tot = tot.sort_values(ascending=False)    #对 tot 进行降序排序，结果保存在 tot 中
tot                                       #统计每一个key出现的次数
```

```python
from lxml import objectify
path = 'examples/Performance_MNR.xml'     # XML文件的路径
parsed = objectify.parse(open(path))     #使用lxml.obejctify解析文件
root = parsed.getroot()    #用getroot获得对xml文件的根节点的引用， 获取XML文件的根节点

data = []    
skip_fields = ['PARENT_SEQ','INDICATOR_SEQ','DESIRED_CHANGE','DECIMAL_PLACES'] # 要跳过的字段列表
for elt in root.INDICATOR:               # 遍历每个INDICATOR节点，
                                        #root.INDICATOR可以返回一个生成器，产生每一个<indicator>的xml元素
    el_data = {}                         # 存储当前节点的数据
    for child in elt.getchildren():       # 遍历当前节点的子节点
        if child.tag in skip_fields:
            continue                       # 跳过需要跳过的字段
        el_data[child.tag] = child.pyval    # 将子节点的标签作为键，值作为值存储到字典中
    data.append(el_data)                   # 将当前节点的数据添加到数据列表中
perf = pd.DataFrame(data)
perf.head()
```

```python
import sqlite3    #驱动生成一个SQLite数据库
query = """
DROP TABLE IF EXISTS test;
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
c REAL, d INTEGER);
"""
con = sqlite3.connect('mydata.sqlite')    #使用 connect() 方法连接到名为 'mydata.sqlite' 的 SQLite 数据库。
con.execute(query)    #执行了一个 SQL 查询语句，其中 query 是你需要执行的 SQL 查询语句。execute() 方法用于执行查询。
con.commit()    #调用 commit() 方法将之前的执行结果提交到数据库中。这是因为 SQLite 数据库默认情况下处于自动提交模式，但是在某些情况下，你可能希望手动提交更改。
```

```python
data = [('atlanta','georgia',1.25,6),('tallahassee','florida',2.6,3),('sacramento','california',1.7,5)]
stmt = "INSERT INTO test VALUES(?,?,?,?)"    #包含四个占位符 ?，分别对应每个值的位置。
con.executemany(stmt,data)    #使用 executemany() 方法执行 SQL 插入语句 stmt，并将列表 data 中的数据进行批量插入
                              #executemany() 方法可以一次性插入多个数据，它会根据占位符的数量将每个元组的值填充到 SQL 语句中。
con.commit()                  #调用 commit() 方法将插入的数据提交到数据库中
```

```python
cursor = con.execute('select * from test')   #使用 con.execute() 方法执行 SELECT * FROM test 查询语句。
                                            #这将返回一个 cursor 游标对象，用于处理查询结果。(py的sql驱动一般是返回元组列表)
rows = cursor.fetchall()     #使用 fetchall() 方法从游标对象（相当于指针）中获取所有查询结果的行，将其存储在变量 rows 中
                              #数据库select的结果只是一个指针，调用指针才能获取指针指向的内容
rows
```

```python
cursor.description    #是一个属性，用于获取查询结果集中每列的描述信息
```

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])   #将数据库转换成frame格式，利于数据分析和处理
```

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],'data2':range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2)    #merge会自动将重叠列名作为连接的键，以列为单位拼接表格，无数据的行用0填充
```

```python
pd.merge(df1,df2,on='key')    #可显式地指定连接键
```

```python
df3 = pd.DataFrame({'lkey':['b','b','a','c','a','a','b'],'data1':range(7)})
df4 = pd.DataFrame({'rkey':['a','b','d'],'data2':range(3)})
pd.merge(df3,df4,left_on='lkey',right_on='rkey')   #连接操作将按照'lkey'列和'rkey'列的值进行匹配
#merge是内连接，结果中的键是两张表的交集，故没有c和d
```

```python
pd.merge(df1,df2,how='outer')   #outer设置外连接，即取两张表并集有c和d，inner是内连接
                                #left指对所有左表的键进行联合，right对所有右表进行联合
```

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','b'],'data1':range(6)})
df2 = pd.DataFrame({'key':['a','b','a','b','d'],'data2':range(5)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='left')   #多对多连接是行的笛卡尔积（进行多对多连接时，每个键的所有可能组合都会被保留下来）
                                        #df1有3个b，df2有2个b，结果行就有6个b
```

```python
pd.merge(df1,df2,how='inner')
```

```python
left = pd.DataFrame({'key1':['foo','foo','bar'],'key2':['one','two','one'],'lval':[1,2,3]})
right = pd.DataFrame({'key1':['foo','foo','bar','bar'],'key2':['one','one','one','two'],'rval':[4,5,6,7]})
pd.merge(left,right,on=['key1','key2'],how='outer')   #以key1和key2为连接键，外连接
```

```python
pd.merge(left,right,on='key1')  #重叠命名的列（不在on中），会以xy区分
```

```python
pd.merge(left,right,on='key1',suffixes=('_left','_right'))  #suffixes是重合列名的命名后缀选项
```

```python
#left              合并时操作中左边的DataFrame
#right             合并时操作中右边的DataFrame
#how              'inner'、'outer'、'left'、'right'之一：默认是'inner'
#on               需要连接的列名。必须是在两边的DataFrame对象都有的列名，并以left和right 中的列名的交集作为连接键
#left_on          left DataFrame中用作连接键的列
#right_on         right DataFrame 中用作连接键的列
#left_index       使用left的行索引作为它的连接键（如果是Multilndex，则是多个键)
#right_index      使用right的行索引作为它的连接键（如果是Multilndex，则是多个键)
#sort         通过连接键按字母顺序对合并的数据进行排序；在默认情况下为True（在大数据集上某些情况下禁用该功能可以获得更好的性能)
#suffixes     在重叠情况下，添加到列名后的字符串元组，默认是（'_x"，’y"）
#copy      如果为False，则在某些特殊情况下避免将数据复制到结果数据结构中；默认情况下总是复制
#indicator      添加一个特殊的列merge，指示每一行的来源，值将根据每行中连接数据的来源分别为‘left_only'，‘right_only'或‘both'
```

```python
#concat函数用于沿指定轴将多个数据集进行简单堆叠。
#它可以按照指定的轴（行或列）将多个数据集堆叠在一起，形成一个大的数据集。
#这种合并不依赖于数据集之间的关系，只是简单地按照指定的轴进行堆叠，没有基于共同的列或索引进行匹配。

#merge函数用于根据共同的列或索引将两个或多个数据集进行合并。
#它可以根据指定的列或索引进行匹配，将共同的行合并到一起，并根据指定的合并方式（如内连接、左连接、右连接等），将非公共的行也合并在一起。
```

```python
data = pd.read_csv('examples/macrodata.csv')
data
```

```python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
#使用pd.PeriodIndex()方法通过指定年份和季度创建一个"PeriodIndex"对象"periods"，其中"year"和"quarter"分别为"data"中的两个列
columns = pd.Index(['realgdp','infl','unemp'],name='item')
#使用pd.Index()方法创建一个"Index"对象"columns"，包含三列数据："realgdp"、"infl"和"unemp"，并且设置该对象的名称为"item"。
data = data.reindex(columns=columns)
#使用data.reindex()方法将"data"重新索引为列由"columns"指定的DataFrame，确保每个时间戳都有对应的三个值
data.index = periods.to_timestamp('D','end')
#将"data"的索引设置为"periods"的时间戳，并使用to_timestamp()方法将它们转换为D频率（每天时间戳），以表示每个季度结束的日期
ldata = data.stack().reset_index().rename(columns={0:'value'})
#使用stack()方法将"data"多重索引的列"realgdp"、"infl"和"unemp"压缩到一个单列"values"中
#使用reset_index()方法将重塑后的DataFrame还原为默认索引形式
#使用rename()方法将重新命名DataFrame的"0"列为"value"
ldata
```

```python
ldata[:10]
```

```python
pivoted = ldata.pivot('date','item','value')  
#使用pivot()方法，指定每一个"date"作为透视表的行索引，每一个"item"作为透视表的列索引，"value"作为透视表填充的值
pivoted
```

```python
ldata['value2'] = np.random.randn(len(ldata))
ldata
```

```python
pivoted = ldata.pivot('date','item')
pivoted[:5]
```

```python
pivoted['value'][:5]
```

```python
unstacked = ldata.set_index(['date','item']).unstack('item')
#pivot方法等价于使用set_index创建分层索引，然后调用unstack
unstacked
```

```python
df = pd.DataFrame({'key':['foo','bar','baz'],'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})
df
```

```python
melted = pd.melt(df,['key'])   #key作为唯一的分组指标，melt是pivot（轴旋转）的反操作(行变纵)
#pd.melt() 方法可以将指定的 DataFrame 对象 df 按照指定的列进行融合，生成一个新的融合后的 DataFrame。
melted
```

```python
reshaped = melted.pivot('key','variable','value')
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A','B'])
#df：要融合的 DataFrame 对象。
#id_vars：用于保持原样的列名或列索引的列表。这些列将保持不变，而不参与融合操作。
#value_vars：要融合的列名或列索引的列表。这些列将被融合在一起，并且将生成一个包含融合值和对应列名的新列。
```

```python
pd.melt(df, value_vars=['A','B','C'])   #只融合abc
```
